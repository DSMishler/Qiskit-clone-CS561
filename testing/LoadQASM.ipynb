```python
from qiskit import *
qc = QuantumCircuit.from_qasm_str("""
OPENQASM 2.0;
include "qelib1.inc";

qreg q[5]; // Quantum Register
creg c[5]; // Classical Register
 
x q[0];
measure q[0] -> c[0];
if (c==1) x q[1];
if (c==1) measure q[1] -> c[1];
""")
qc.draw('mpl')
```

```python
backend = BasicAer.get_backend('qasm_simulator')
execute(qc,backend).result().get_counts()
```
